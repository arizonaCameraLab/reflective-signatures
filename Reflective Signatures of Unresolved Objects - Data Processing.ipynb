# Import code Dependencies

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Lab Data Processing

## Validating Forward Model

```python
datapath1 = "./familyP1_target.csv"
data1 = pd.read_csv(datapath1).to_numpy()[:,1]

datapath2 = "./familyP2_target.csv"
data2= pd.read_csv(datapath2).to_numpy()[:,1]

fig, (ax, ax2) = plt.subplots(1,2,figsize = (20,10), layout = 'constrained')

ax.plot(data1, linewidth = 3)
ax.grid()
ax.set_xlabel("Frame #", fontsize = 25, fontweight="bold")
ax.set_ylabel("$V_i$: Mean Code Value of Target ROI", fontsize = 25, fontweight="bold")
ax.set_title("(a) Experimentally Acquired $S_G$ for Surface $P_1(x)$", fontsize = 23, fontweight="bold")
ax.tick_params(axis='both', which='major', labelsize=15)
ax.tick_params(axis='both', which='major', labelsize=20)

ax2.plot(data2, linewidth = 3)
ax2.grid()
ax2.set_xlabel("Frame #", fontsize = 25, fontweight="bold")
ax2.set_ylabel("$V_i$: Mean Code Value of Target ROI", fontsize = 25, fontweight="bold")
ax2.set_title("(b) Experimentally Acquired $S_G$ for Surface $P_2(x)$", fontsize = 23, fontweight="bold")
ax2.tick_params(axis='both', which='major', labelsize=15)
ax2.tick_params(axis='both', which='major', labelsize=20)
```

## Additional Indoor Lab Experiments

```python
target1fast = pd.read_csv("./target_1_fast.csv").to_numpy()[:,1]
target1fastreversed = pd.read_csv("./target_1_fast_reversed.csv").to_numpy()[:,1]

target2fast = pd.read_csv("./target_2_fast.csv").to_numpy()[:,1]
target2fastreversed = pd.read_csv("./target_2_fast_reversed.csv").to_numpy()[:,1]
target2slow = pd.read_csv("./target_2_slow.csv").to_numpy()[:,1]

target3fast = pd.read_csv("./target_3_fast.csv").to_numpy()[:,1]
```
